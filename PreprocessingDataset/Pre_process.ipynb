```python
import cv2
import dlib
from PIL import Image
import face_alignment
from skimage import io
import torch
import os
```

```python

# Initialize dlib's face detector and face alignment model (FAN)
detector = dlib.get_frontal_face_detector()
fa = face_alignment.FaceAlignment(face_alignment.LandmarksType.TWO_D, device='cuda' if torch.cuda.is_available() else 'cpu')
```

```python
def align_face(image_path):
    # Load the image
    image = io.imread(image_path)
    
    # Detect face in the image
    dets = detector(image, 1)
    
    if len(dets) == 0:
        print(f"No face found in {image_path}")
        return None
    
    # Get the first face detected
    face = dets[0]
    
    # Get landmarks using FAN
    landmarks = fa.get_landmarks(image)
    
    if landmarks is None:
        print(f"No landmarks found for {image_path}")
        return None
    
    # Align the image based on the landmarks
    aligned_image = align_and_crop(image, landmarks[0])
    return aligned_image
```

```python
def align_and_crop(image, landmarks):
    # Crop the face based on landmarks (this function can be adjusted for better cropping)
    # Here we crop around the facial features, adjusting size as necessary for StyleGAN2 input.
    face = Image.fromarray(image)
    # Convert to desired size (say, 256x256 for StyleGAN2)
    resized_face = face.resize((256, 256))
    return resized_face
```

```python
directory = '/home/sharjeel/dev/myenv1/projects/customStylegan2adapytorch/Dataset/img_align_celeba'
```

```python
for imgFile in os.listdir(directory):
    if imgFile.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff', '.gif')):
        img_path = os.path.join(directory, imgFile)
        aligned_face = align_face(img_path)
        if aligned_face:
            aligned_face.save(img_path)
        else:
            if os.path.isfile(img_path):
                os.remove(img_path)
                print(f"Deleted {img_path}")
            else:
                print(f"File {img_path} does not exist.")
```
